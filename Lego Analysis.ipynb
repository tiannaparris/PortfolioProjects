```python
# Importing the library I will use for this project

import pandas as pd
```

```python
# Reading in the data

df = pd.read_csv('https://raw.githubusercontent.com/KeithGalli/lego-analysis/master/datasets/lego_sets.csv')
parent_theme = pd.read_csv('https://raw.githubusercontent.com/KeithGalli/lego-analysis/master/datasets/parent_themes.csv')
```

```python
# Looking at the data
```

```python
# Dataset - Lego Sets

df.head()
```

```python
# Dataset - Parent Theme

parent_theme.head()
```

```python
# Merging the data sets

merged = df.merge(parent_theme, left_on='parent_theme', right_on='name')

merged.head()
```

```python
# Removing duplicate columns

merged = merged.drop(columns='name_y')

merged.head()
```

```python
# Checking for missing values

merged[merged['set_num'].isnull()].shape
```

```python
# Task 1
```

```python
# What percentage of all licensed sets ever released were Star Wars themed?
```

```python
# Filtering the data for only licensed sets

licensed = merged[merged['is_licensed']]
licensed = licensed.dropna(subset=['set_num'])

licensed.head()
```

```python
# Filtering for all licensed sets ever released were Star Wars themed

star_wars = licensed[licensed['parent_theme']=='Star Wars']

star_wars.head()
```

```python
# Calculating percentage 

percent_theme = int(star_wars.shape[0]/licensed.shape[0]*100)

print(percent_theme,'%')
```

```python
# 51% of all licensed sets ever released were Star Wars themed.
```

```python
# Task 2
```

```python
# In which year was Star Wars not the most popular licensed theme?
```

```python
sorted_licensed = licensed.sort_values('year')
sorted_licensed['count'] = 1

df_sum = sorted_licensed.groupby(['year','parent_theme']).sum().reset_index()
df_max = df_sum.sort_values('count', ascending = False).drop_duplicates(['year'])

df_max.sort_values('year', inplace=True)
df_max
```

```python
# In 2017, Super Heroes dethroned Star Wars as the most popular theme.
```
